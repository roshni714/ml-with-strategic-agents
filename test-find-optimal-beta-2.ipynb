```python
from agent_distribution import AgentDistribution
from utils import compute_continuity_noise, compute_contraction_noise, compute_score_bounds
from optimal_beta import optimal_beta_empirical_policy_loss, optimal_beta_expected_policy_loss
from utils import fixed_point_interpolation_true_distribution
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_types = 1
d = 2
etas = np.random.uniform(3., 8., n_types * d).reshape(n_types, d, 1)
gammas = np.random.uniform(0.05, 2., n_types *d).reshape(n_types, d, 1)
dic = {"etas":etas, "gammas":gammas}
agent_dist = AgentDistribution(n=1000, d=d, n_types=n_types, types=dic, prop=None)

sigma = compute_continuity_noise(agent_dist) + 0.05
q = 0.7

name_exp = "heterogenous_pi3_exp"
true_beta = np.array([np.cos(np.pi/3), np.sin(np.pi/3)]).reshape(2, 1)
f = fixed_point_interpolation_true_distribution(agent_dist, sigma, q, plot=True, savefig="figs/fp_{}.pdf".format(name_exp))
empirical_min_loss, empirical_beta, empirical_s_beta, thetas, empirical_losses = optimal_beta_empirical_policy_loss(agent_dist, sigma, q, f, true_beta=true_beta, plot=True, savefig=None)
expected_min_loss, expected_beta, expected_s_beta, thetas, expected_losses = optimal_beta_expected_policy_loss(agent_dist, sigma, f, true_beta=true_beta, plot=True, savefig=None)
```

```python
print("Empirical:", empirical_min_loss, empirical_beta, empirical_s_beta)
print("Expcted:", expected_min_loss, expected_beta, expected_s_beta)
```

```python
plt.plot(thetas, empirical_losses, label="empirical loss")
plt.plot(thetas, expected_losses, label="expected loss")
plt.ylabel("Loss")
plt.xlabel("Theta (Corresponds to Beta)")
plt.title("L(beta) vs. beta")
plt.legend()
plt.savefig("figs/loss_{}.pdf".format(name_exp))
```
